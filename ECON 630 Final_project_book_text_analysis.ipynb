### Corpus Creation Description
To create the dataset of NYT bestselling non-fiction books the NYT API has been used. The weekly list were scraped from 2014/01/12 - 2023/08/06. Since the bestseller list only includes a one-line description of the book, the dataset has been extended to include the long description of the book. The longer description of the books was scraped via the Google Books API using the isbn number as the unique identifier from the initial NYT list. Last but not least more information about the authors were collected via the Wikimedia API by scraping the infobox of the author's wikipedia page, if it exists.

#### Corpus Caveats
The New York Times Non-Fiction Best Seller list is a significant and influential indicator of the popularity of non-fiction books in the United States. It reflects the books that are selling well and capturing the attention of readers across the country. However, it's important to note that while the list provides valuable insights into book sales and trends, it may not be a perfect representation of the absolute most widely read books for several reasons:

- Limited Sample Size: The list is based on a specific sample of bookstores, both independent and chain retailers, as well as wholesalers across the United States. While this sample is substantial, it may not capture all sales channels, such as online retailers, libraries, or direct author sales.

- Regional Variations: Book sales can vary significantly based on regional preferences and demographics. The New York Times list might not fully capture regional best-sellers or reading habits.

- Prominence of Physical Sales: The list traditionally emphasizes physical book sales over digital or e-book sales. With the rise of e-books and digital reading platforms, this could lead to an underrepresentation of certain segments of the reading population.

- Marketing and Promotion: Best-seller lists can be influenced by marketing efforts, author tours, media coverage, and promotional campaigns. This means that books with strong marketing backing might have a higher chance of making it to the list, even if they might not reflect the reading preferences of the broader population.

- Time Lag: The list is updated weekly, which might not fully capture the long-term popularity of certain books that have a more gradual buildup in sales over time.

- Subjectivity of Curation: The list is curated by editors at The New York Times, and their selection process can involve subjective judgment. Some books that are critically acclaimed but might not fit neatly into the categories defined by the list could be left out.

#### Improvements opportunities to the corpus (not implemented due to time constraint)
While the New York Times Non-Fiction Best Seller list provides valuable insights into popular non-fiction books, it's important to complement this information with other sources of data and research methods to get a more comprehensive understanding of the most widely read books in the United States. Other sources, such as Amazon best-seller rankings, Goodreads popularity, online reviews, and surveys, can provide additional perspectives on popular reading habits and preferences.

### Research Question 1: What are the dominant topics in US non-fiction literature over the past ten years? How do those topics change over time?

To start this analysis I will create word clouds per year on both the short description of the books from the NYT as well as the longer description that was scraped via the Google Books API (the text you find describing the book across all e-commerce platforms).

```python
from gensim.models import TfidfModel
from gensim.matutils import corpus2csc
from sklearn.feature_extraction.text import CountVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import gensim
from gensim import corpora, models
```

```python
from collections import OrderedDict

# Get the unique years from the 'year' column
unique_years = df_books['list_year'].unique()

# Create sub DataFrames for each unique year
sub_dataframes = OrderedDict()  # Ordered dictionary to store sub DataFrames

for year in sorted(unique_years):
    sub_df = df_books[df_books['list_year'] == year]
    sub_dataframes[year] = sub_df
```

```python
def generate_wordcloud_from_df(dataframe, column_name, custom_stopwords=None, top_n=20, font_path=None):
    # Convert the specified column to a list of strings
    corpus = dataframe[column_name].astype(str).tolist()

    # Get the default English stopwords from sklearn
    default_stopwords = set(CountVectorizer(stop_words='english').get_stop_words())
    
    # Combine default 'english' stopwords and custom stopwords
    if custom_stopwords:
        combined_stopwords = list(default_stopwords.union(custom_stopwords))
    else:
        combined_stopwords = list(default_stopwords)

    # Initialize the CountVectorizer
    vectorizer = CountVectorizer(stop_words=combined_stopwords)

    # Fit and transform the corpus into a bag of words representation
    X = vectorizer.fit_transform(corpus)

    # Get the feature names (words)
    feature_names = vectorizer.get_feature_names_out()

    # Convert the sparse matrix to a dense array for better readability
    dense_array = X.toarray()

    # Calculate the sum of word occurrences across all documents
    word_frequencies = dense_array.sum(axis=0)

    # Create a mapping of word to frequency
    word_freq_mapping = dict(zip(feature_names, word_frequencies))

    # Sort the words based on frequency and get the top words
    top_words = sorted(word_freq_mapping, key=word_freq_mapping.get, reverse=True)[:top_n]

    # Create a WordCloud using the specified font
    if font_path:
        wordcloud = WordCloud(width=800, height=400, background_color='white', font_path=font_path).generate_from_frequencies({word: word_freq_mapping[word] for word in top_words})
    else:
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies({word: word_freq_mapping[word] for word in top_words})

    # Display the WordCloud using matplotlib
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
# Loop over the dictionary of DataFrames
for df_name, df in sub_dataframes.items():
    print(df_name)
    custom_stopwords = ['draws','including','history','book', 'time', 'just','readers','bestseller', 'years','best','author', 'like', 'year','way', 'new','times', 'bestselling','york','story','world', '10', 'describes']
    generate_wordcloud_from_df(df, 'summary', custom_stopwords=custom_stopwords, top_n=15, font_path="/Library/Fonts/Arial Unicode.ttf")
```

### Topic Modeling

https://www.analyticsvidhya.com/blog/2016/08/beginners-guide-to-topic-modeling-in-python/

There are many approaches for obtaining topics from a text such as – Term Frequency and Inverse Document Frequency. NonNegative Matrix Factorization techniques. Latent Dirichlet Allocation is the most popular topic modeling technique and in this article, we will discuss the same.

LDA assumes documents are produced from a mixture of topics. Those topics then generate words based on their probability distribution. Given a dataset of documents, LDA backtracks and tries to figure out what topics would create those documents in the first place.

LDA is a matrix factorization technique. In vector space, any corpus (collection of documents) can be represented as a document-term matrix. The following matrix shows a corpus of N documents D1, D2, D3 … Dn and vocabulary size of M words W1,W2 .. Wn. The value of i,j cell gives the frequency count of word Wj in Document Di.

```python
import pandas as pd
import string
from gensim import corpora, models
from gensim.parsing.preprocessing import STOPWORDS
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

# APPENDIX

```python
def perform_topic_modeling(df):
    # Copy the original DataFrame
    df_copy = df.copy()
    # Convert the 'summary' column to lowercase
    df_copy['summary'] = df_copy['summary'].str.lower()

    # Define custom stopwords
    custom_stopwords = ['•','american','book','life', '#1', 'time', 'just', 'readers', 'bestseller', 'years', 'best', 'author', 'like', 'year', 'way', 'new', 'times', 'bestselling', 'york', 'story', 'world']

    # Combine custom stopwords and gensim stopwords
    all_stopwords = set(STOPWORDS).union(custom_stopwords)

    # Drop rows with missing or empty text
    df_copy.dropna(subset=['summary'], inplace=False)
    df_copy = df_copy[df_copy['summary'] != '']

    # Preprocessing: Tokenize and create a dictionary
    documents = df_copy['summary'].tolist()
    translator = str.maketrans('', '', string.punctuation)
    tokenized_documents = [doc.translate(translator).split() for doc in documents]

    # Remove stopwords
    filtered_documents = [[word for word in doc if word not in all_stopwords] for doc in tokenized_documents]
    dictionary = corpora.Dictionary(filtered_documents)

    # Create a bag-of-words representation
    corpus = [dictionary.doc2bow(doc) for doc in filtered_documents]

    # Combine all documents into a single document
    combined_document = [word for doc in filtered_documents for word in doc]

    # Create a bag-of-words representation for the combined document
    combined_corpus = [dictionary.doc2bow(combined_document)]

    # Perform LDA topic modeling on the combined document
    num_topics = 1  # Specify the number of topics
    num_words_per_topic = 25 # Print the topics and their top words
    lda_model = models.LdaModel(combined_corpus, num_topics=num_topics, id2word=dictionary, passes=20)

    # Print the topics and their top words
    for idx, topic in lda_model.print_topics(-1):
        words = topic.split("+")
        words_only = [word.split("*")[1].strip() for word in words][:num_words_per_topic]
        print(f"Topic {idx}: {' '.join(words_only)}")
```

```python
for df_name, df in sub_dataframes.items():
    # Example usage:
    print(df_name)
    perform_topic_modeling(df)
```
